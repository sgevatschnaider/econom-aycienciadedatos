```python
from IPython.display import display, HTML

html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Herramientas y Técnicas Econométricas de la Revolución de la Credibilidad</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            line-height: 1.8;
            background-color: #f9f9f9;
            margin: 20px;
            padding: 20px;
        }
        h1 {
            color: #2c3e50;
            text-align: center;
            font-size: 2.5em;
            margin-bottom: 20px;
        }
        h2 {
            color: #2c3e50;
            font-size: 1.8em;
            margin-top: 20px;
        }
        p {
            font-size: 1.2em;
            margin-bottom: 15px;
        }
        ul {
            list-style-type: disc;
            margin-left: 30px;
            margin-bottom: 20px;
        }
        li {
            margin-bottom: 10px;
            font-size: 1.1em;
        }
        .highlight {
            color: #d35400;
            font-weight: bold;
        }
    </style>
</head>
<body>
    <h1>Herramientas y Técnicas Econométricas de la Revolución de la Credibilidad</h1>
    <p>
        Las herramientas y técnicas econométricas utilizadas en la <span class="highlight">"revolución de la credibilidad"</span> en la economía empírica están enfocadas en mejorar la identificación de relaciones causales.
        A continuación se presentan algunas de las principales herramientas econométricas empleadas:
    </p>

    <h2>Ensayos Controlados Aleatorizados (RCTs)</h2>
    <p>
        Los RCTs son considerados el estándar de oro para establecer causalidad. Al asignar aleatoriamente a los sujetos a grupos de tratamiento y control, ayudan a eliminar el sesgo de selección y los factores de confusión.
    </p>

    <h2>Variables Instrumentales (IV)</h2>
    <p>
        Las técnicas de IV se utilizan cuando la aleatorización no es posible. Un instrumento es una variable que está correlacionada con el tratamiento pero no está correlacionada con el término de error en la ecuación del resultado. Esto ayuda a abordar la endogeneidad y el sesgo de variables omitidas.
    </p>

    <h2>Diferencias en Diferencias (DiD)</h2>
    <p>
        DiD es una técnica cuasi-experimental que compara los cambios en los resultados a lo largo del tiempo entre un grupo de tratamiento y un grupo de control. Ayuda a controlar la heterogeneidad no observada que es constante en el tiempo.
    </p>

    <h2>Diseños de Discontinuidad de Regresión (RDD)</h2>
    <p>
        RDD se utiliza cuando el tratamiento se asigna en función de un punto de corte. Al comparar individuos justo por encima y por debajo del umbral, los investigadores pueden estimar el efecto local del tratamiento.
    </p>

    <h2>Métodos de Emparejamiento</h2>
    <p>
        Los métodos de emparejamiento, como el emparejamiento por puntajes de propensión, emparejan individuos en los grupos de tratamiento y control con características observables similares, lo que ayuda a controlar las variables de confusión en contextos no aleatorizados.
    </p>

    <h2>Modelos de Efectos Fijos</h2>
    <p>
        Los modelos de efectos fijos controlan la heterogeneidad no observada al tener en cuenta características individuales que no varían en el tiempo. Esto se utiliza comúnmente en análisis de datos de panel para aislar los efectos causales.
    </p>

    <h2>Método de Control Sintético</h2>
    <p>
        Este método se utiliza para estimar el impacto causal de intervenciones en estudios de caso comparativos, construyendo una combinación ponderada de unidades de control que mejor se aproxima a las características de la unidad tratada antes de la intervención.
    </p>

    <h2>Experimentos Naturales</h2>
    <p>
        Estos explotan shocks externos o exógenos (como cambios en políticas o eventos naturales) que afectan solo a parte de la población, imitando una asignación aleatoria.
    </p>

    <h2>Triple Diferencias (DDD)</h2>
    <p>
        Este método extiende el enfoque de diferencias en diferencias al agregar otra dimensión, lo que permite analizar los efectos causales a través de tres grupos o períodos de tiempo.
    </p>

    <h2>Análisis de Datos de Panel y Longitudinales</h2>
    <p>
        Estas técnicas permiten a los investigadores rastrear a los mismos individuos o entidades a lo largo del tiempo, ayudando a controlar factores no observables que no varían en el tiempo.
    </p>

    <h2>Métodos Bayesianos</h2>
    <p>
        La econometría bayesiana proporciona un marco flexible para incorporar información previa en el proceso de estimación, lo que la hace particularmente útil en el contexto de modelos complejos o conjuntos de datos pequeños.
    </p>

    <h2>Diseño de Kinks en Regresión (RKD)</h2>
    <p>
        Este método es similar a RDD pero se enfoca en cambios en la pendiente de la función de regresión en lugar de una discontinuidad en un punto de corte.
    </p>

    <h2>Modelos de Selección de Heckman</h2>
    <p>
        Estos modelos se utilizan para corregir el sesgo de selección, especialmente cuando la muestra de datos observada no es aleatoria, sino que está sujeta a una selección basada en el resultado de interés.
    </p>

    <h2>Método Generalizado de Momentos (GMM)</h2>
    <p>
        GMM es un método de estimación flexible utilizado en modelos de datos de panel dinámicos, donde técnicas tradicionales como MCO pueden estar sesgadas debido a la endogeneidad.
    </p>

    <h2>Regresión por Cuantiles</h2>
    <p>
        Esta técnica permite a los investigadores estimar el impacto de las variables en diferentes puntos de la distribución de resultados, no solo en la media, proporcionando una imagen más completa del efecto del tratamiento.
    </p>

    <h2>Puntaje de Propensión con Balance de Covariables (CBPS)</h2>
    <p>
        Esta es una mejora respecto a los métodos tradicionales de emparejamiento por puntajes de propensión, y tiene como objetivo equilibrar mejor las covariables entre los grupos de tratamiento y control para mejorar la inferencia causal.
    </p>

    <h2>Simulaciones de Monte Carlo</h2>
    <p>
        Estas se utilizan para evaluar las propiedades de los estimadores en modelos econométricos simulando datos bajo diferentes suposiciones, ayudando a comprender la robustez de los resultados.
    </p>

    <p>
        Estas herramientas representan el cambio hacia métodos de investigación empírica más sólidos, transparentes y creíbles, característicos de la revolución de la credibilidad en la economía. El énfasis está en estrategias claras de identificación y en minimizar los sesgos, asegurando que los resultados empíricos sean confiables y replicables.
    </p>
</body>
</html>
"""

# Mostrar el contenido HTML con estilo
display(HTML(html_content))
```

```python
from IPython.display import display, HTML

html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Ensayos Controlados Aleatorizados (RCTs)</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            line-height: 1.8;
            background-color: #f9f9f9;
            margin: 20px;
            padding: 20px;
        }
        h1 {
            color: #2c3e50;
            text-align: center;
            font-size: 2.5em;
            margin-bottom: 20px;
        }
        h2 {
            color: #2c3e50;
            font-size: 1.8em;
            margin-top: 20px;
        }
        p {
            font-size: 1.2em;
            margin-bottom: 15px;
            background-color: #ffffff;
            padding: 10px;
            border-radius: 5px;
            box-shadow: 0 0 10px rgba(0, 0, 0, 0.1);
        }
        ul {
            list-style-type: disc;
            margin-left: 30px;
            margin-bottom: 20px;
        }
        li {
            margin-bottom: 10px;
            font-size: 1.1em;
        }
        .highlight {
            color: #d35400;
            font-weight: bold;
        }
        .container {
            margin-bottom: 20px;
        }
    </style>
</head>
<body>
    <h1>Ensayos Controlados Aleatorizados (RCTs)</h1>

    <div class="container">
        <p>
            Los <span class="highlight">Ensayos Controlados Aleatorizados (RCTs, por sus siglas en inglés)</span> son un método experimental ampliamente utilizado para identificar efectos causales de manera rigurosa. Este enfoque se basa en la
            <span class="highlight">aleatorización</span> de una intervención o tratamiento, lo que significa que los participantes se asignan al azar a un grupo de tratamiento o a un grupo de control. Esta aleatorización ayuda a eliminar el
            <span class="highlight">sesgo de selección</span>, ya que asegura que las diferencias entre los grupos se deben únicamente al tratamiento y no a otros factores no observados.
        </p>
        <p>
            Los RCTs se consideran el estándar de oro en la investigación experimental, especialmente en disciplinas como la medicina, pero su uso en economía ha crecido significativamente en las últimas décadas, especialmente en el campo de la
            <span class="highlight">economía del desarrollo</span> y la <span class="highlight">economía laboral</span>.
        </p>
    </div>

    <h2>Proceso de los RCTs</h2>
    <div class="container">
        <ul>
            <li><strong>Selección de la muestra:</strong> Se comienza seleccionando una población de interés que será dividida en dos grupos: el grupo de tratamiento (que recibirá la intervención) y el grupo de control (que no la recibirá).</li>
            <li><strong>Aleatorización:</strong> La aleatorización es el corazón del RCT. Cada participante tiene la misma probabilidad de ser asignado al grupo de tratamiento o al de control. Esto asegura que ambos grupos sean similares en términos de características observables y no observables.</li>
            <li><strong>Aplicación del tratamiento:</strong> Al grupo de tratamiento se le aplica la intervención o política que se desea evaluar (ej. una nueva política educativa, un programa de subsidios), mientras que el grupo de control no recibe dicha intervención.</li>
            <li><strong>Medición de los resultados:</strong> Se comparan los resultados de ambos grupos, y cualquier diferencia se puede atribuir causalmente al tratamiento.</li>
            <li><strong>Análisis de los datos:</strong> Los datos se analizan utilizando técnicas estadísticas que comparan los resultados promedio entre los grupos de tratamiento y control para estimar el <span class="highlight">efecto causal promedio</span>.</li>
        </ul>
    </div>

    <h2>Ejemplo en Economía</h2>
    <div class="container">
        <p>
            Supongamos que un gobierno está interesado en evaluar el impacto de un <span class="highlight">programa de transferencia monetaria</span> en los ingresos de los hogares. El gobierno selecciona una muestra de 1,000 hogares y aleatoriamente asigna
            500 hogares al grupo de tratamiento, que recibirán una transferencia mensual, mientras que los otros 500 forman el grupo de control, que no recibirá nada. Al final de un año, los investigadores comparan el cambio en los ingresos promedio
            entre los dos grupos. Si los hogares que recibieron la transferencia tienen ingresos significativamente más altos, se puede concluir que el programa tuvo un impacto positivo.
        </p>
    </div>

    <h2>Ventajas de los RCTs</h2>
    <div class="container">
        <ul>
            <li><strong>Alta validez interna:</strong> Gracias a la aleatorización, cualquier diferencia en los resultados puede atribuirse con confianza al tratamiento, otorgando una alta validez interna.</li>
            <li><strong>Eliminación del sesgo de selección:</strong> En estudios observacionales, los individuos pueden seleccionar si participan en un programa o no, lo que puede sesgar los resultados. Los RCTs eliminan este sesgo.</li>
            <li><strong>Similitud con experimentos naturales:</strong> Los RCTs replican los estándares de los experimentos controlados en las ciencias naturales, haciendo que los resultados sean más creíbles.</li>
        </ul>
    </div>

    <h2>Desafíos de los RCTs</h2>
    <div class="container">
        <ul>
            <li><strong>Costos y logística:</strong> Implementar un RCT puede ser costoso y logísticamente complicado, especialmente en contextos amplios o en países en desarrollo.</li>
            <li><strong>Problemas éticos:</strong> En ocasiones, es poco ético asignar tratamientos de manera aleatoria, especialmente si se trata de intervenciones que podrían tener un impacto significativo en la vida de los participantes.</li>
            <li><strong>Validez externa:</strong> Aunque los RCTs tienen una alta validez interna, sus resultados pueden no ser generalizables a otras poblaciones o contextos.</li>
            <li><strong>Efecto Hawthorne:</strong> Existe el riesgo de que los participantes cambien su comportamiento simplemente por el hecho de estar involucrados en un experimento.</li>
            <li><strong>Limitación temporal:</strong> Los RCTs generalmente se enfocan en los efectos de corto plazo de una intervención, lo que puede no captar los impactos a largo plazo.</li>
        </ul>
    </div>

    <h2>Importancia de los RCTs en Economía</h2>
    <div class="container">
        <p>
            Los RCTs han revolucionado el campo de la economía empírica, particularmente en áreas como la <span class="highlight">economía del desarrollo</span> y la <span class="highlight">economía laboral</span>.
            <strong>Esther Duflo, Abhijit Banerjee, y Michael Kremer</strong>, ganadores del <strong>Premio Nobel de Economía en 2019</strong>, han sido pioneros en el uso de RCTs para evaluar intervenciones políticas en países en desarrollo.
            Han estudiado el impacto de las intervenciones en educación, salud y pobreza mediante RCTs, proporcionando evidencia rigurosa sobre qué políticas funcionan mejor para combatir la pobreza.
        </p>
    </div>

    <h2>Conclusión</h2>
    <div class="container">
        <p>
            Los <span class="highlight">Ensayos Controlados Aleatorizados (RCTs)</span> son una herramienta poderosa para identificar relaciones causales en economía. Al asegurar que la única diferencia sistemática entre los grupos de tratamiento y control
            es la intervención, los RCTs permiten a los economistas evaluar de manera confiable el impacto de políticas y programas. Aunque presentan desafíos en términos de costos y generalización, su capacidad para eliminar el sesgo de selección y
            proporcionar pruebas claras los convierte en un estándar en la investigación económica empírica.
        </p>
    </div>

</body>
</html>
"""

# Mostrar el contenido HTML con estilo en Google Colab
display(HTML(html_content))
```

```python
# Simular una población
np.random.seed(42)
n = 1000
X = np.random.rand(n)  # Covariable (por ejemplo, nivel de ingresos)
treatment = np.random.binomial(1, 0.5, size=n)  # Asignar tratamiento aleatoriamente

# Definir resultados potenciales
Y0 = 2 + 3 * X + np.random.randn(n)  # Resultados del grupo de control
Y1 = Y0 + 5  # El efecto del tratamiento suma 5 unidades

# Resultados observados basados en la asignación de tratamiento
Y_observed = np.where(treatment == 1, Y1, Y0)

# Crear DataFrame
df = pd.DataFrame({
    'X': X,
    'treatment': treatment,
    'Y_observed': Y_observed
})

# Diccionarios para colores y etiquetas
colors = {0: 'skyblue', 1: 'salmon'}
labels = {0: 'Control', 1: 'Tratamiento'}
```

```python
def animate_scatter():
    fig, ax = plt.subplots(figsize=(10, 6))
    sample_sizes = np.linspace(50, n, 50, dtype=int)

    def update(num):
        ax.clear()
        current_n = sample_sizes[num]
        df_sample = df.iloc[:current_n]

        for treatment_group in df_sample['treatment'].unique():
            subset = df_sample[df_sample['treatment'] == treatment_group]
            ax.scatter(subset['X'], subset['Y_observed'], c=colors[treatment_group],
                       label=labels[treatment_group], alpha=0.6)
            # Ajustar una línea de regresión si hay suficientes puntos
            if len(subset) > 1:
                coeffs = np.polyfit(subset['X'], subset['Y_observed'], deg=1)
                x_vals = np.linspace(0, 1, 100)
                y_vals = coeffs[0] * x_vals + coeffs[1]
                ax.plot(x_vals, y_vals, color=colors[treatment_group], linestyle='--')

        ax.set_title(f'Resultados Observados vs X (n={current_n})', fontsize=16)
        ax.set_xlabel('X (Covariable)', fontsize=12)
        ax.set_ylabel('Y_observed', fontsize=12)
        ax.legend(title='Grupo')
        ax.grid(True)
        ax.set_xlim(0, 1)
        ax.set_ylim(df['Y_observed'].min() - 1, df['Y_observed'].max() + 1)

    ani = animation.FuncAnimation(fig, update, frames=len(sample_sizes), repeat=False)
    plt.close(fig)
    return ani

# Crear y mostrar la animación
ani3 = animate_scatter()
HTML(ani3.to_jshtml())
```

```python
def animate_histogram():
    fig, ax = plt.subplots(figsize=(10, 6))
    sample_sizes = np.linspace(50, n, 50, dtype=int)
    bins = np.linspace(df['Y_observed'].min(), df['Y_observed'].max(), 30)

    def update(num):
        ax.clear()
        current_n = sample_sizes[num]
        df_sample = df.iloc[:current_n]

        for treatment_group in df_sample['treatment'].unique():
            subset = df_sample[df_sample['treatment'] == treatment_group]
            ax.hist(subset['Y_observed'], bins=bins, alpha=0.5, label=labels[treatment_group],
                    color=colors[treatment_group], density=True)

        ax.set_title(f'Distribución de Y_observed por Grupo (n={current_n})', fontsize=16)
        ax.set_xlabel('Y_observed', fontsize=12)
        ax.set_ylabel('Densidad', fontsize=12)
        ax.legend(title='Grupo')
        ax.grid(True)
        ax.set_xlim(df['Y_observed'].min(), df['Y_observed'].max())
        ax.set_ylim(0, None)

    ani = animation.FuncAnimation(fig, update, frames=len(sample_sizes), repeat=False)
    plt.close(fig)
    return ani

# Crear y mostrar la animación
ani4 = animate_histogram()
HTML(ani4.to_jshtml())
```

```python
def animate_boxplot():
    fig, ax = plt.subplots(figsize=(8, 6))
    sample_sizes = np.linspace(50, n, 50, dtype=int)

    def update(num):
        ax.clear()
        current_n = sample_sizes[num]
        df_sample = df.iloc[:current_n]

        df_sample.boxplot(column='Y_observed', by='treatment', widths=0.6, patch_artist=True,
                          boxprops=dict(facecolor='skyblue', color='blue'),
                          medianprops=dict(color='red'), ax=ax)
        ax.set_title(f'Boxplot de Y_observed por Grupo (n={current_n})', fontsize=16)
        ax.set_xlabel('Grupo', fontsize=12)
        ax.set_ylabel('Y_observed', fontsize=12)
        ax.set_xticklabels(['Control', 'Tratamiento'])
        plt.suptitle('')
        ax.grid(True)
        ax.set_ylim(df['Y_observed'].min() - 1, df['Y_observed'].max() + 1)

    ani = animation.FuncAnimation(fig, update, frames=len(sample_sizes), repeat=False)
    plt.close(fig)
    return ani

# Crear y mostrar la animación
ani5 = animate_boxplot()
HTML(ani5.to_jshtml())
```
